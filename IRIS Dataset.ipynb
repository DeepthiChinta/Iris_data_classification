# Manual Modelling
By looking at data and deciding some cutoff points for modelling

Look at Relational plot of petal width Vs petal Length
We can observe that if petal length < 2.5(say) then it is Setosa...Else if petal length < 4.5 then it is Versicolor Else it is Virginca

```python
def single_feature_prediction(petal_length):
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.5:
        return 1
    else:
        return 2
```

```python
manual_predictions = np.array([single_feature_prediction(val) for val in X_train[:,2]])
# : is to specify all values and 2 is indicating that we need to take second column values i.e,petal length 
```

```python
manual_predictions
```

```python
Y_train
```

```python
manual_predictions == Y_train
```

```python
manual_model_accuracy = np.mean(manual_predictions == Y_train) # 90% we accuarately classified correctly
```

```python
print(f"Manual model accuracy : {manual_model_accuracy * 100:.2f}%")
```

### Using cross-validation to evaluate our model 

Cross validation is a technique used in machine learning to evaluate the performance of a model on unseen data. It involves dividing the available data into multiple folds or subsets, using one of these folds as a validation set, and training the model on the remaining folds. This process is repeated multiple times, each time using a different fold as the validation set. Finally, the results from each validation step are averaged to produce a more robust estimate of the modelâ€™s performance.

<img src="https://scikit-learn.org/stable/_images/grid_search_cross_validation.png" width="500" height="500">

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
accuracies = cross_val_score(model,X_train,Y_train,cv=5,scoring="accuracy") # atlast fold we can see that accucay is 1 and remaining it is 95%
```

```python
np.mean(accuracies)  # Before it is 96.67 but it is 97.5
```

We can seee 2 differenecs in them

```python
def plot_incorrect_predictions(df_predictions,x_axis_feature,y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10,10) )
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_predictions", data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    
    plt.show()
```

```python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
```
